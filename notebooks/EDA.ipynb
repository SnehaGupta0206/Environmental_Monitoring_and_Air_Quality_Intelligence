```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df["AQI"].hist(bins=30)
plt.title("Distribution of AQI")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
city_aqi = df.groupby("City")["AQI"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
city_aqi.head(10).plot(kind="bar")
plt.title("Top 10 Cities by Average AQI")
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.show()
```

```python
pollutants = ["PM2.5", "PM10", "NO", "NO2", "NOx", "NH3", "CO", "SO2", "O3"]

df[pollutants].mean().sort_values().plot(kind="bar", figsize=(10,5))

plt.title("Average Pollutant Levels")
plt.xlabel("Pollutants")
plt.ylabel("Average Concentration")
plt.xticks(rotation=45)
plt.show()
```

Observation: PM2.5 and PM10 show comparatively higher average concentrations than several other pollutants, indicating that particulate matter is a major contributor to air pollution. This suggests that controlling particulate emissions should be a priority for improving air quality.

```python
monthly_aqi = df.groupby("Month")["AQI"].mean()

plt.figure(figsize=(10,5))
monthly_aqi.plot(marker="o")
plt.title("Average Monthly AQI")
plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.grid(True)
plt.show()
```

Observation

The graph shows how the average AQI changes from month to month. Some months have higher AQI values, indicating poorer air quality, while others have lower AQI values. This suggests that air pollution varies throughout the year and may be influenced by seasonal weather conditions and human activities.

```python
print("Average AQI:", round(df["AQI"].mean(), 2))
print("Maximum AQI:", df["AQI"].max())
print("Minimum AQI:", df["AQI"].min())
print("Number of Cities:", df["City"].nunique())
```

Observation

The KPI values provide a quick summary of the dataset. They show the overall average AQI, the highest and lowest AQI recorded, and the number of cities included in the analysis. These indicators help understand the overall air quality status of the dataset.
